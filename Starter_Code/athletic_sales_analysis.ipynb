```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.

df_2020 = athletic2020 = pd.read_csv ('Resources/athletic_sales_2020.csv')
df_2021 = athletic2021 = pd.read_csv ('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
df_2020.head()
```

```python
# Display the 2021 sales DataFrame
df_2021.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
df_2020.info()
```

```python
# Check the 2021 sales data types.
df_2021.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales = pd.concat([df_2020, df_2021], ignore_index=True)
combined_sales.reset_index(drop=True, inplace=True)
combined_sales.head()
```

```python
# Check if any values are null.
combined_sales.isnull().any()
```

```python
# Check the data type of each column
combined_sales.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales['invoice_date'] = pd.to_datetime(combined_sales['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_sales.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
product_sold_summary = combined_sales.groupby(['region', 'state', 'city'], as_index=False)['units_sold'].sum()
product_sold_summary.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
products_sold_sorted = product_sold_summary.sort_values(by='Total_Products_Sold', ascending=False).head(5)
products_sold_sorted.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_table = pd.pivot_table(combined_sales, 
                             values='units_sold',
                             index = ['region', 'state', 'city'], 
                             aggfunc = 'sum').reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_table.rename(columns= {'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
pivot_table_sorted = pivot_table.sort_values(by='Total_Products_Sold', ascending=False).head(5)
pivot_table_sorted.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
regional_sales = combined_sales.groupby(['region','state','city',])['total_sales'].sum().reset_index()
regional_sales.rename(columns={'total_sales':'Total_Sales'}, inplace=True)

# Show the top 5 results.
regional_sales_sorted = regional_sales.sort_values(by='Total_Sales', ascending=False).head(5)
regional_sales_sorted.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_table_regional = pd.pivot_table(combined_sales, 
                             values='total_sales',
                             index = ['region', 'state', 'city'], 
                             aggfunc = 'sum').reset_index()  

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table_regional.rename(columns= {'total_sales': 'Total_Sales'}, inplace=True)
pivot_table_regional_sorted = pivot_table_regional.sort_values(by='Total_Sales', ascending=False)


# Show the top 5 results.
pivot_table_regional_sorted.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
retailer_sales = combined_sales.groupby(['retailer','region','state','city',])['total_sales'].sum().reset_index()
retailer_sales.rename(columns={'total_sales':'Total_Sales'}, inplace=True)

# Show the top 5 results.
retailer_sales_sorted = retailer_sales.sort_values(by='Total_Sales', ascending=False).head(5)
retailer_sales_sorted.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
pivot_table_retail = pd.pivot_table(combined_sales, 
                             values='total_sales',
                             index = ['retailer', 'region', 'state', 'city'], 
                             aggfunc = 'sum').reset_index()  

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table_retail.rename(columns= {'total_sales': 'Total_Sales'}, inplace=True)
pivot_table_retail_sorted = pivot_table_retail.sort_values(by='Total_Sales', ascending=False).head(5)

# Show the top 5 results.
pivot_table_retail_sorted.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_athletic_footwear=combined_sales[combined_sales['product'] == "Women's Athletic Footwear"]
womens_athletic_footwear.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_sales = womens_athletic_footwear.groupby(['retailer','region', 'state','city',])['units_sold'].sum().reset_index()
women_sales.rename(columns={'units_sold':'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the top 5 results.
womens_sales_sorted = women_sales.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)
womens_sales_sorted.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_table_ws=pd.pivot_table(womens_athletic_footwear, 
                             values='units_sold',
                             index = ['retailer','region', 'state', 'city'], 
                             aggfunc = 'sum').reset_index()  

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivot_table_ws.rename(columns= {'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the top 5 results.
pivot_table_ws_sorted = pivot_table_ws.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)
pivot_table_ws_sorted.head()
```
